```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

T = np.linspace(0, 10, 16)
x_blue = [1, 1.2, 0.9, 1.1, 1, 1, 1.1, 0.8, 1, 1.1, 1.3, 1.5, 2, 1.9, 1.8, 2]
x_red = [1.6, 1.3, 1.2, 1, 0.8, 0.9, 1, 1.8, 1.7, 1.8, 1.75, 1.9, 1.65, 1.7, 1.75, 1.9]

fig, axs = plt.subplots(1, 1, figsize=(14, 5))
axs.scatter(T, x_blue, marker='s', label="2")
axs.scatter(T, x_red, label="1")
N = 100
xmin, xmax = min(T), max(T)
xx = np.linspace(xmin, xmax, N)

t, c, k = interpolate.splrep(T, x_blue, s=1, k=4)
spline = interpolate.BSpline(t, c, k, extrapolate=False)
axs.plot(xx, spline(xx), 'b')

t, c, k = interpolate.splrep(T, x_red, s=1, k=4)
spline = interpolate.BSpline(t, c, k, extrapolate=False)
axs.plot(xx, spline(xx), 'r')

# disabling xticks by Setting xticks to an empty list
axs.set_xticks([])

# disabling yticks by setting yticks to an empty list
axs.set_yticks([])
axs.legend()
axs.set_xlim([T[0], T[-1]])

axs.set_xlabel(r'$t_{N-1}$', fontsize=15)
axs.set_ylabel(r'$x_i(t)$', fontsize=15)

# Add 4 equally spaced dotted grey vertical lines
for i in [1, 2, 3]:
    x_val = (1/4) * i * T[-1]
    axs.axvline(x=x_val, color='grey', linestyle='dotted')

plt.show()
```
